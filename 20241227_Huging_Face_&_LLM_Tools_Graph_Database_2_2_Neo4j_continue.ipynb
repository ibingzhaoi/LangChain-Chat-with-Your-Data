

RecursiveCharacterTextSplitter：是一個文本分割器，用於將長文本按照指定規則分割成更小的片段。  
is_separator_regex=False：表示分割器不使用正則表達式作為分隔符，而是使用默認的邏輯進行分割。

在 `RecursiveCharacterTextSplitter` 中，`is_separator_regex` 是一個布爾型參數，用於指定分隔文本時，分隔符是否應該被解釋為正則表達式 (regex)。

當 `is_separator_regex = False` 時：

1. **分隔符被視為普通字符串：**
   - 分隔符直接被用作一個普通的字符串來進行匹配，而不是一個正則表達式模式。
   - 此時，分割行為不會支持正則表達式的靈活匹配功能，比如字符集、量詞、組合模式等。

2. **更簡單的匹配方式：**
   - 適合用於靜態且明確的分隔符，例如句點 (`.`)、換行符 (`\n`)、空格 (` `) 等。
   - 如果您知道文本的結構且分隔符是固定的，這是一個較高效且易於理解的選項。

3. **性能考量：**
   - 因為不使用正則表達式，匹配分隔符的過程會更快速，尤其是在處理非常大的文本時。

````python
def split_form10k_data_from_file(file):
    chunks_with_metadata = [] # use this to accumlate chunk records
    file_as_object = json.load(open(file)) # open the json file
    for item in ['item1','item1a','item7','item7a']: # pull these keys from the json
        print(f'Processing {item} from {file}')
        item_text = file_as_object[item] # grab the text of the item
        item_text_chunks = text_splitter.split_text(item_text) # split the text into chunks
        chunk_seq_id = 0
        for chunk in item_text_chunks[:20]: # only take the first 20 chunks
            form_id = file[file.rindex('/') + 1:file.rindex('.')] # extract form id from file name
            # finally, construct a record with metadata and the chunk text
            chunks_with_metadata.append({
                'text': chunk,
                # metadata from looping...
                'f10kItem': item,
                'chunkSeqId': chunk_seq_id,
                # constructed metadata...
                'formId': f'{form_id}', # pulled from the filename
                'chunkId': f'{form_id}-{item}-chunk{chunk_seq_id:04d}',
                # metadata from file...
                'names': file_as_object['names'],
                'cik': file_as_object['cik'],
                'cusip6': file_as_object['cusip6'],
                'source': file_as_object['source'],
            })
            chunk_seq_id += 1
        print(f'\tSplit into {chunk_seq_id} chunks')
    return chunks_with_metadata
````

這段代碼是用來生成一個唯一的 `chunkId`，其格式結合了多個元素，用於唯一標識某個文本片段。以下是詳細解析：

---

### **代碼解釋**

```python
'chunkId': f'{form_id}-{item}-chunk{chunk_seq_id:04d}'
```

#### **使用了 Python 的 f-string 格式化**
- f-string 是一種格式化字符串的方式，可以在字符串中直接插入變量或運行表達式，語法為 `f'{expression}'`。

---

#### **組成部分**

1. **`{form_id}`**
   - 插入變量 `form_id` 的值。
   - 這通常是某個表單的 ID，例如從文件名中提取的部分。

2. **`{item}`**
   - 插入變量 `item` 的值。
   - 這可能是某個 JSON 鍵的名稱，例如 `item1` 或 `item7`，用來指示片段屬於表單中的哪個部分。

3. **`chunk`**
   - 固定字符串 `chunk`，用來表示這是一個文本片段。

4. **`{chunk_seq_id:04d}`**
   - 插入變量 `chunk_seq_id` 的值，並應用了格式化規則 `:04d`。
   - **`04d` 的含義：**
     - `0`：如果數字不足 4 位，則用前置零補齊。
     - `4`：確保數字至少占用 4 位寬度。
     - `d`：表示數字格式。

   - 例如：
     - 當 `chunk_seq_id = 1` 時，結果為 `0001`。
     - 當 `chunk_seq_id = 23` 時，結果為 `0023`。

---

### **生成的結果**
假設變量的值如下：
```python
form_id = 'form12345'
item = 'item1'
chunk_seq_id = 7
```

生成的 `chunkId` 為：
```plaintext
'form12345-item1-chunk0007'
```

---

### **用途**

1. **唯一標識片段**：
   - 通過結合表單 ID（`form_id`）、表單項目（`item`）和片段序號（`chunk_seq_id`），`chunkId` 確保每個片段在整個數據集中都是唯一的。

2. **便於檢索和管理**：
   - `chunkId` 可以用於檢索特定的文本片段，因為它包含了片段的上下文信息（來自哪個表單、哪個項目，以及序號）。

3. **數據存儲一致性**：
   - 在將數據存儲到 Neo4j 或其他資料庫時，`chunkId` 作為主鍵，確保不會插入重複的數據。

---

### **示例應用場景**

#### **文本分割與存儲**
假設您處理了一份 Form 10-K 文件，提取了 `item1` 並分割為片段：
- `form_id` 是文件名 `form12345`。
- 每個片段依次分配 `chunk_seq_id`，從 0 開始。

生成的 `chunkId` 示例：
```plaintext
form12345-item1-chunk0000
form12345-item1-chunk0001
form12345-item1-chunk0002
```

#### **用於數據查詢**
存儲到資料庫後，您可以根據 `chunkId` 快速查找特定片段，例如：
- 查找屬於 `form12345` 並對應 `item1` 的第 7 個片段：
  ```plaintext
  form12345-item1-chunk0007
  ```


這段代碼的作用是將一個文本片段及其元數據作為一個節點存儲到 Neo4j 資料庫中。如果該節點已存在（基於 `chunkId` 唯一標識符），則不創建新節點；否則，創建一個新節點並設置相關屬性。以下是詳細解釋：

---

### MERGE

#### **Cypher 查詢**
```cypher
MERGE(mergedChunk:Chunk {chunkId: $chunkParam.chunkId})
    ON CREATE SET
        mergedChunk.names = $chunkParam.names,
        mergedChunk.formId = $chunkParam.formId,
        mergedChunk.cik = $chunkParam.cik,
        mergedChunk.cusip6 = $chunkParam.cusip6,
        mergedChunk.source = $chunkParam.source,
        mergedChunk.f10kItem = $chunkParam.f10kItem,
        mergedChunk.chunkSeqId = $chunkParam.chunkSeqId,
        mergedChunk.text = $chunkParam.text
RETURN mergedChunk
```

1. **`MERGE(mergedChunk:Chunk {chunkId: $chunkParam.chunkId})`**
   - 這行代碼的作用是讓 Neo4j 嘗試查找資料庫中是否已存在一個符合條件的節點。
   - 如果找到匹配的節點，`MERGE` 停止後續的創建邏輯。
   - `MERGE` 嘗試在資料庫中查找是否已存在一個具有 `chunkId` 等於 `$chunkParam.chunkId` 的節點。
   - 如果找到該節點，則直接返回該節點。
   - 如果未找到，則創建一個類型為 `Chunk` 的新節點，並使用 `ON CREATE SET` 設置其屬性。
   - 是否執行 `ON CREATE SET` 的判斷邏輯由 `MERGE` 自動完成，無需顯式指定。
   - `MERGE` 確保不會創建重複的節點。
   - `MERGE` 是一個高級子句，將查找和創建的行為封裝在一起。
   - 用戶不需要顯式指定「如果找到就返回」「如果未找到就創建」，這些行為是由 `MERGE` 的內部機制自動執行的。

2. **`ON CREATE SET`**
   - 僅在創建新節點時執行。
   - 將 `$chunkParam` 字典中的值設置為 `mergedChunk` 節點的屬性。

   **設置的屬性包括：**
   - `names`：可能是公司名稱列表。
   - `formId`：表單 ID，例如從文件名中提取。
   - `cik`：公司識別碼（CIK）。
   - `cusip6`：公司證券標識符的前 6 位。
   - `source`：數據來源，例如文件來源或提交來源。
   - `f10kItem`：對應於 Form 10-K 的具體部分（如 `item1`）。
   - `chunkSeqId`：文本片段的序列號。
   - `text`：文本片段的實際內容。

3. **`RETURN mergedChunk`**
   - 返回被 `MERGE` 查找到或創建的節點 `mergedChunk`。

---

#### **Python 調用**
```python
kg.query(merge_chunk_node_query,
         params={'chunkParam': first_file_chunks[0]})
```

1. **查詢執行**
   - 使用 `kg.query` 將 Cypher 查詢語句發送到 Neo4j 資料庫。

2. **參數傳遞**
   - 通過 `params` 傳遞查詢所需的參數。
   - `first_file_chunks[0]` 是一個字典，包含文本片段的所有元數據，假設其結構如下：

     ```python
         "chunkSeqId": 0,
         "text": "This is the text of the chunk..."
     }
     ```

### **執行邏輯**
1. **檢查節點是否存在**
   - 如果資料庫中已存在具有 `chunkId` 為 `"unique-id-001"` 的節點，則直接返回該節點，`ON CREATE SET` 不會執行。

2. **創建新節點**
   - 如果資料庫中沒有該節點，則創建一個新節點並設置屬性。
   - 使用 `ON CREATE SET` 將 `$chunkParam` 中的值存儲到節點的屬性中。

3. **返回結果**
   - 返回查找到或新創建的節點，用於驗證操作是否成功。

### CONSTRAINT

這段代碼使用了 Neo4j 的 Cypher 查詢語法來創建一個唯一性約束，以下是詳細解釋：

---

### **代碼解析**

```cypher
CREATE CONSTRAINT unique_chunk IF NOT EXISTS
    FOR (c:Chunk) REQUIRE c.chunkId IS UNIQUE
```

1. **`CREATE CONSTRAINT`**：
   - 用於創建約束，以確保資料庫中的數據滿足特定條件。
   - 這裡創建的是唯一性約束，確保在所有 `Chunk` 節點中，`chunkId` 屬性是唯一的。

2. **`unique_chunk`**：
   - 約束的名稱是 `unique_chunk`，便於後續管理和查看該約束。
   - 命名建議使用描述性名稱，幫助開發者理解約束的用途。

3. **`IF NOT EXISTS`**：
   - 如果約束已存在，則不會重複創建，這樣可以避免運行時出現錯誤。
   - 提高查詢的安全性，適合在重複執行創建查詢的腳本中使用。

4. **`FOR (c:Chunk)`**：
   - 指定約束適用於所有具有 `Chunk` 標籤的節點。

5. **`REQUIRE c.chunkId IS UNIQUE`**：
   - 定義約束的具體條件，要求每個 `Chunk` 節點的 `chunkId` 屬性值在整個資料庫中唯一。

---

### **功能與作用**

1. **唯一性保障**：
   - 確保資料庫中的 `Chunk` 節點不會有重複的 `chunkId`。
   - 有效避免因多次插入相同數據而導致的重複節點問題。

2. **加速查詢**：
   - 唯一性約束通常會隱式地創建一個索引，提升基於 `chunkId` 的查詢性能。

3. **數據一致性**：
   - 確保 `chunkId` 是 `Chunk` 節點的唯一標識符，有助於維護數據的邏輯完整性。

---

### **執行後的影響**

- **如果約束不存在**：
  - 創建一個新的唯一性約束。
  - 未來的插入或更新操作將受到此約束的限制，禁止插入重複的 `chunkId` 值。

- **如果約束已存在**：
  - 不執行任何操作，保持現有的唯一性約束。

---

### **查看約束**

執行以下查詢可以查看資料庫中的所有約束，包括 `unique_chunk`：

```cypher
SHOW CONSTRAINTS
```

結果將列出約束的詳細信息，例如名稱、類型、適用的節點和屬性。

---

### **常見錯誤與解決方法**

1. **插入重複值錯誤**：
   - 當您嘗試插入具有相同 `chunkId` 的節點時，資料庫會拋出錯誤。
   - 解決方法：在插入數據前，使用 `MERGE` 或檢查是否已存在該節點。

2. **刪除約束**：
   - 如果需要刪除此約束，可以使用以下查詢：
     ```cypher
     DROP CONSTRAINT unique_chunk
     ```

---

### **應用場景**

1. **數據唯一性保證**：
   - 適用於需要唯一標識符的節點，例如用於文本分割片段（`Chunk`）。

2. **索引與性能優化**：
   - 提升基於 `chunkId` 的查詢效率，例如查找特定的分割片段。

---

### **總結**

這段代碼的主要作用是創建一個唯一性約束，確保 `Chunk` 節點的 `chunkId` 屬性值在資料庫中唯一。同時，它提高了數據一致性和查詢性能，是管理和保護核心數據的重要措施。

`UNWIND` 是 Neo4j Cypher 查詢語言中用於處理列表的一個關鍵字，它的作用是將列表中的每個元素「展開」（即依次取出），並作為單獨的值進行後續處理。以下是對 `UNWIND $chunks AS chunk` 的詳細解釋：

---

### **句法結構**
```cypher
UNWIND $chunks AS chunk
```

1. **`$chunks`**：
   - 這是一個傳遞給查詢的參數，必須是一個列表。
   - 列表中的每個元素代表一個數據記錄，通常是類似字典的結構，例如：
     ```json
     [
         {"chunkId": "chunk-001", "text": "First chunk text"},
         {"chunkId": "chunk-002", "text": "Second chunk text"}
     ]
     ```

2. **`AS chunk`**：
   - 將列表中的每個元素逐一取出，並賦值給變量 `chunk`。
   - 這個變量可以在後續的查詢中使用，表示當前正在處理的列表元素。

---

### **UNWIND 的作用**

#### 1. 將列表元素拆分成多行
當您傳遞一個列表給 Cypher 查詢時，`UNWIND` 的作用是將列表中的每個元素分別作為單獨的一行處理。這非常適合用於批量處理數據。

#### 2. 與其他語句配合使用
`UNWIND` 通常與其他 Cypher 子句（如 `MERGE`、`CREATE` 或 `SET`）配合使用，來對展開的每個元素進行操作。

1. **批量插入數據**
   - 當您需要一次插入多個記錄（例如來自 JSON 文件或 API 的數據）時，`UNWIND` 非常有用。

2. **處理列表屬性**
   - 如果某個節點屬性是一個列表，您可以使用 `UNWIND` 將列表中的每個值分解為單獨的節點或關係。

3. **高效執行查詢**
   - 將所有數據打包成一個列表傳遞給查詢，然後使用 `UNWIND` 進行批量處理，可以減少多次查詢的開銷。


### **`$` 符號**

在 Neo4j 的 Cypher 查詢語言中，`$` 符號用於引入**參數化查詢**。這是一種將查詢中的值分離為外部變量的方法，用於提高查詢的安全性、性能以及靈活性。

### **`$` 的作用**

1. **表示參數的佔位符**：
   - 在 Cypher 查詢中，`$` 後面跟著一個變量名，表示一個參數的佔位符。
   - 實際的值並不在查詢內部，而是通過外部（如程式碼）動態傳遞給查詢。

2. **與數據庫交互時傳遞值**：
   - 程式語言（如 Python）在執行查詢時，可以將變量的值通過參數傳遞給 Cypher 查詢。
   - 例如，在 Python 中：
     ```python
     kg.query("""
     UNWIND $chunks AS chunk
     MERGE (c:Chunk {chunkId: chunk.chunkId})
     """, params={"chunks": my_chunks})
     ```

   - 這裡的 `$chunks` 是查詢的佔位符，而 `params={"chunks": my_chunks}` 指定了 `$chunks` 的實際值。

---

### **為什麼使用 `$`**

1. **提高查詢安全性**
   - 使用參數化查詢可以防止 SQL/Cypher 注入攻擊，因為參數值是安全地轉義後處理的。

2. **提升性能**
   - 當查詢結構固定而僅參數不同時，資料庫可以重用查詢的執行計劃，而不需要每次重新解析查詢。

3. **靈活性**
   - 可以輕鬆地對同一查詢套用不同的參數，從而適應多種查詢需求。

- 與普通的 `RetrievalQAChain` 相比，`WithSourcesChain` 的輸出包括：
  - **答案 (answer)**：由 LLM（大語言模型）生成。
  - **來源文檔 (source_documents)**：從檢索器（retriever）中獲取的相關文檔，提供支持生成答案的背景資料。

---

### **`.from_chain_type` 的意義**
`.from_chain_type` 是一個**工廠方法**，用於快速生成一個 `RetrievalQAWithSourcesChain` 實例。通過這個方法，可以使用不同的鏈（Chain）類型，例如 `"stuff"`、`"map_reduce"` 或 `"refine"`，來定義如何將檢索到的文檔與查詢結合並生成答案。

#### **為什麼用 `.from_chain_type`？**
- **簡化初始化流程**：
  - 用戶只需指定 `chain_type` 和相關參數，而不需要顯式構建整個鏈的細節。
- **靈活性**：
  - 支持多種文檔處理方式，適應不同的場景需求。
- **一致性**：
  - 讓初始化方式標準化，便於切換和擴展。

---

### **工作原理**
以下是 `RetrievalQAWithSourcesChain.from_chain_type` 的工作邏輯：

1. **定義 LLM（大語言模型）**
   - 傳入一個支持 LangChain 的 LLM 實例，如 `ChatOpenAI`。
   - 此 LLM 負責生成答案。

2. **指定檢索器（Retriever）**
   - 檢索器負責根據查詢從數據庫（如向量存儲）中檢索相關文檔。

3. **指定鏈類型（Chain Type）**
   - 決定如何處理檢索到的文檔並結合查詢生成答案：
     - **`"stuff"`**：將所有檢索到的文檔拼接後，直接交給 LLM。
     - **`"map_reduce"`**：分步處理文檔，每步生成部分答案，最後合併。
     - **`"refine"`**：逐步改進答案，每次處理一個文檔並基於前一次結果進行完善。

4. **返回一個完整的 RAG 問答鏈**
   - 初始化後，可以直接用於查詢，並同時返回答案和來源。


---

### **`WithSourcesChain` 與來源文檔**

`WithSourcesChain` 的主要特性是：
1. **答案透明性**：
   - 提供來源文檔，讓生成的答案更可信。
   - 適用於需要引用數據來源的應用場景，例如法律、醫療或研究領域。

2. **返回來源文檔的格式**：
   - 每個來源文檔通常包含元數據（如來源標籤、文檔 ID）和內容。
   - 格式示例：
     ```python
     response["source_documents"] = [
         {"content": "Climate change is accelerating...", "metadata": {"source": "doc1"}},
         {"content": "Mitigation efforts are critical...", "metadata": {"source": "doc2"}}
     ]
     ```


####`chain_type`

---

### **1. `stuff`**
- **描述**：
  - 最簡單的類型，直接將檢索到的文檔內容拼接起來作為 LLM 的輸入。
  - 適合處理少量的檢索結果，避免超過模型的輸入限制。

- **適用場景**：
  - 檢索的文檔數量較少，且每個文檔內容簡短。
  - 需要快速處理和生成答案。

- **優勢**：
  - 簡單高效，適合處理小型數據集。

- **限制**：
  - 無法處理大量檢索結果，容易超過 LLM 的上下文限制（例如 GPT 的 token 限制）。

---

### **2. `map_reduce`**
- **描述**：
  - 分兩步執行：
    1. **Map 步驟**：對每個檢索到的文檔單獨生成部分答案。
    2. **Reduce 步驟**：將所有部分答案結合，生成最終答案。
  - 適合處理大規模檢索結果。

- **適用場景**：
  - 檢索結果數量多，或單個文檔內容較長。
  - 需要模型對每個文檔進行逐一處理並總結。

- **優勢**：
  - 能處理更多的檢索結果，分步生成答案。

- **限制**：
  - 比 `stuff` 慢，對於小數據集可能過於複雜。

---

### **3. `refine`**
- **描述**：
  - 逐步生成答案：
    1. 初始回答基於第一個檢索文檔生成。
    2. 每個後續文檔用於逐步改進之前的答案。
  - 類似於反覆完善答案的過程。

- **適用場景**：
  - 對答案準確性要求較高，且需要利用所有檢索到的內容。
  - 答案可能需要結合多個文檔的信息。

- **優勢**：
  - 適合複雜問題和需要多文檔融合的情境。
  - 最終答案能充分利用檢索結果。

- **限制**：
  - 效率較低，隨檢索文檔數量增加，計算成本顯著提高。

---

### **4. 自定義類型**
- **描述**：
  - 您可以設計自己的 `chain_type`，滿足特定的業務需求。
  - 通常通過 `chain_type_kwargs` 傳遞額外參數。

- **適用場景**：
  - 現有類型無法滿足需求，或需要特殊的處理邏輯。

---

### **選擇合適的 `chain_type`**

| **`chain_type`** | **適用場景**                     | **優勢**                     | **限制**                       |
|------------------|--------------------------------|----------------------------|------------------------------|
| `stuff`          | 小型檢索結果，簡單問答         | 高效，實現簡單              | 無法處理大量檢索結果         |
| `map_reduce`      | 大量檢索結果，長文檔處理       | 支持大規模處理              | 相對較慢                     |
| `refine`          | 高準確性需求，多文檔融合       | 答案逐步完善，準確性高      | 計算成本高，效率較低         |
| 自定義類型        | 特定場景或業務需求             | 完全定制化                 | 需要額外開發與調試           |

### `prettychain`

這段代碼定義了一個名為 `prettychain` 的函數，用於優雅地打印問答鏈（chain）的回應。以下是對這段代碼的詳細解析：

---

### **函數功能**
- **主要目的**：
  - 對給定的問題（`question`），使用問答鏈（`chain`）生成答案。
  - 將答案按固定的行寬格式化，進行美觀的打印。

- **輸入參數**：
  - `question: str`：用戶的提問，作為查詢。

- **輸出**：
  - 無返回值，直接在終端中打印格式化的答案。

---

### **代碼解析**

#### **函數定義**
```python
def prettychain(question: str) -> str:
    """Pretty print the chain's response to a question"""
```
- **`question: str`**：函數接受一個字符串類型的參數 `question`，這是用戶的提問。
- **Docstring**：描述函數用途，表示該函數會將問答鏈的回應美觀地打印。

---

#### **生成答案**
```python
response = chain({"question": question},
    return_only_outputs=True,)
```
1. **`chain({"question": question})`**：
   - 調用預先定義的問答鏈 `chain`，將用戶的問題作為輸入。
   - `chain` 是一個 RetrievalQA 或 RetrievalQAWithSourcesChain 類型的對象，負責處理問題，檢索相關內容，並生成答案。

2. **`return_only_outputs=True`**：
   - 這是一個參數，指定只返回查詢結果的輸出部分，而不包括中間的其他數據。
   - 返回值是一個字典，通常包含：
     - `response['answer']`：生成的答案文本。
     - （如果設置 `return_source_documents=True`）可能還會包含 `response['source_documents']`。

---

#### **格式化輸出**
```python
print(textwrap.fill(response['answer'], 60))
```
1. **`response['answer']`**：
   - 從返回的字典中提取答案文本。

2. **`textwrap.fill`**：
   - 使用 Python 的 `textwrap` 模塊格式化文本，將長行文字自動換行為指定寬度。
   - **參數說明**：
     - `response['answer']`：要格式化的文本。
     - `60`：行寬限制，表示每行最多顯示 60 個字符，超過的部分自動換行。

3. **`print`**：
   - 將格式化後的文本輸出到終端。

```python
        else:
            print("No answer found for the given question.")
    except Exception as e:
        print(f"An error occurred: {str(e)}")
```

### Create a linked list of Chunk nodes for each section
- Start by identifying chunks from the same section

````python
cypher = """
  MATCH (from_same_form:Chunk)
    WHERE from_same_form.formId = $formIdParam
  RETURN from_same_form {.formId, .f10kItem, .chunkId, .chunkSeqId } as chunkInfo
    LIMIT 10
"""

kg.query(cypher, params={'formIdParam': form_info['formId']})
````

這段查詢的作用是：

1. **查找節點**：匹配所有 `Chunk` 節點，且屬性 `formId` 等於指定的參數 `$formIdParam`。
2. **返回屬性**：提取匹配節點的部分屬性：`formId`、`f10kItem`、`chunkId`、`chunkSeqId`，作為 `chunkInfo` 返回。
3. **限制結果數量**：僅返回最多 10 個節點。

- Order chunks by their sequence ID

````python
cypher = """
  MATCH (from_same_form:Chunk)
    WHERE from_same_form.formId = $formIdParam
  RETURN from_same_form {.formId, .f10kItem, .chunkId, .chunkSeqId } as chunkInfo
    ORDER BY from_same_form.chunkSeqId ASC
    LIMIT 10
"""

kg.query(cypher, params={'formIdParam': form_info['formId']})
````

這段查詢的作用是：

1. **查找節點**：匹配所有 `Chunk` 節點，且其屬性 `formId` 等於參數 `$formIdParam`。
2. **返回屬性**：提取匹配節點的部分屬性：`formId`、`f10kItem`、`chunkId`、`chunkSeqId`，作為 `chunkInfo` 返回。
3. **排序**：按照 `chunkSeqId` 屬性**升序**排序。
  - ORDER BY XXX ASC 按照 `XXX` 屬性**升序**排序。
  - ASC（升序）
  - DESC（降序）



- Limit chunks to just the "Item 1" section, the organize in ascending order

````python
cypher = """
  MATCH (from_same_section:Chunk)
  WHERE from_same_section.formId = $formIdParam
    AND from_same_section.f10kItem = $f10kItemParam // NEW!!!
  RETURN from_same_section { .formId, .f10kItem, .chunkId, .chunkSeqId }
    ORDER BY from_same_section.chunkSeqId ASC
    LIMIT 10
"""

kg.query(cypher, params={'formIdParam': form_info['formId'],
                         'f10kItemParam': 'item1'})
````

新增條件 `AND from_same_section.f10kItem = $f10kItemParam // NEW!!!` 的效果是：

- **作用**：在查詢 `Chunk` 節點時，新增一個過濾條件，僅返回屬性 `f10kItem` 等於指定參數 `$f10kItemParam` 的節點。
- **結果**：篩選出同時符合 `formId` 和 `f10kItem` 條件的節點。



- Collect ordered chunks into a list

````python
cypher = """
  MATCH (from_same_section:Chunk)
  WHERE from_same_section.formId = $formIdParam
    AND from_same_section.f10kItem = $f10kItemParam
  WITH from_same_section { .formId, .f10kItem, .chunkId, .chunkSeqId }
    ORDER BY from_same_section.chunkSeqId ASC
    LIMIT 10
  RETURN collect(from_same_section) // NEW!!!
"""

kg.query(cypher, params={'formIdParam': form_info['formId'],
                         'f10kItemParam': 'item1'})
````

**`RETURN collect(from_same_section)` 的作用**

1. **`collect(from_same_section)`**：
   - 將 `from_same_section` 中選出的所有節點結果（最多 10 個，按 `chunkSeqId` 排序）收集為一個列表。
   - 返回的格式是單一列表，而不是多行獨立的結果。

2. **與之前的差別**：
   - **之前**：
     ```cypher
     RETURN from_same_section { .formId, .f10kItem, .chunkId, .chunkSeqId }
     ```
     - 每個節點作為單獨一行返回，結果是一個多行結構。
   - **現在**：
     ```cypher
     RETURN collect(from_same_section)
     ```
     - 所有節點結果被收集到一個列表中，結果是一個單行的結構，列表中包含所有符合條件的節點。


- **主要差異**：`collect` 將多行結果整合為單一列表，改變的是返回格式，而非圖數據。
- **圖數據影響**：`RETURN` 或 `collect` 均不改變圖數據結構。

### Add a NEXT relationship between subsequent chunks
- Use the `apoc.nodes.link` function from Neo4j to link ordered list of `Chunk` nodes with a `NEXT` relationship
- Do this for just the "Item 1" section to start

````python
cypher = """
  MATCH (from_same_section:Chunk)
  WHERE from_same_section.formId = $formIdParam
    AND from_same_section.f10kItem = $f10kItemParam
  WITH from_same_section
    ORDER BY from_same_section.chunkSeqId ASC
  WITH collect(from_same_section) as section_chunk_list
    CALL apoc.nodes.link(
        section_chunk_list,
        "NEXT",
        {avoidDuplicates: true}
    )  // NEW!!!
  RETURN size(section_chunk_list)
"""

kg.query(cypher, params={'formIdParam': form_info['formId'],
                         'f10kItemParam': 'item1'})
````

### **`CALL apoc.nodes.link` 的功能**

`CALL apoc.nodes.link` 是 Neo4j APOC 庫中提供的一個便利方法，用於在一組節點之間快速創建關係。這段代碼的作用是：

1. **創建關係**：
   - 為 `section_chunk_list` 中的節點逐一創建關係。
   - 關係類型為 `"NEXT"`，表示這些節點之間存在一種有序的連接。

2. **參數解釋**：
   - **`section_chunk_list`**：
     - 這是一個節點的列表（由 `collect` 收集而成）。
     - 這些節點是查詢篩選後的 `Chunk` 節點，按 `chunkSeqId` 排序。
   - **`"NEXT"`**：
     - 關係的類型名稱，表示在節點之間創建 `:NEXT` 關係。
     - 可以替換為其他類型名稱（如 `"LINKED_TO"` 或 `"FOLLOWS"`）。
   - **`{avoidDuplicates: true}`**：
     - 避免重複創建已存在的 `:NEXT` 關係，確保資料庫中不會有多條重複的邊。
APOC 是 "Awesome Procedures On Cypher" 的縮寫，是 Neo4j 提供的一組擴展庫，包含大量的預定義存儲過程（procedures）和功能（functions），用於補充 Cypher 查詢語言的功能。它幫助用戶完成一些常見但較複雜的圖數據操作。

````python
cypher = """
  MATCH (from_same_section:Chunk)
  WHERE from_same_section.formId = $formIdParam
    AND from_same_section.f10kItem = $f10kItemParam
  WITH from_same_section
    ORDER BY from_same_section.chunkSeqId ASC
  WITH collect(from_same_section) as section_chunk_list
    CALL apoc.nodes.link(
        section_chunk_list,
        "NEXT",
        {avoidDuplicates: true}
    )
  RETURN size(section_chunk_list)
"""
for form10kItemName in ['item1', 'item1a', 'item7', 'item7a']:
  kg.query(cypher, params={'formIdParam':form_info['formId'],
                           'f10kItemParam': form10kItemName})
````

### Information is stored in the structure of a graph
- Matched patterns of nodes and relationships in a graph are called **paths**
- The length of a path is equal to the number of relationships in the path
- Paths can be captured as variables and used elsewhere in queries

````python
cypher = """
    MATCH window = (c1:Chunk)-[:NEXT]->(c2:Chunk)-[:NEXT]->(c3:Chunk)
        WHERE c1.chunkId = $chunkIdParam
    RETURN length(window) as windowPathLength
    """

kg.query(cypher,
         params={'chunkIdParam': next_chunk_info['chunkId']})
````

### Finding variable length windows
- A pattern match will fail if the relationship doesn't exist in the graph
- For example, the first chunk in a section has no preceding chunk, so the next query won't return anything

````python
cypher = """
    MATCH window=(c1:Chunk)-[:NEXT]->(c2:Chunk)-[:NEXT]->(c3:Chunk)
        WHERE c2.chunkId = $chunkIdParam
    RETURN nodes(window) as chunkList
    """
# pull the chunk ID from the first
kg.query(cypher,
         params={'chunkIdParam': first_chunk_info['chunkId']})
````

first_chunk 沒有之前的 chunk ，返回空集合。如果我們希望對找到的段落找出一段 窗口資料，那我們要處理可能查到的段落正好是第一段或最後一段的情況。

- Modify `NEXT` relationship to have variable length

````python
cypher = """
  MATCH window=
      (:Chunk)-[:NEXT*0..1]->(c:Chunk)-[:NEXT*0..1]->(:Chunk)
    WHERE c.chunkId = $chunkIdParam
  RETURN length(window)
  """

kg.query(cypher,
         params={'chunkIdParam': first_chunk_info['chunkId']})
````

這裡允許沒有前一段或後一段的 patent match

- Retrieve only the longest path

````python
cypher = """
  MATCH window=
      (:Chunk)-[:NEXT*0..1]->(c:Chunk)-[:NEXT*0..1]->(:Chunk)
    WHERE c.chunkId = $chunkIdParam
  WITH window as longestChunkWindow
      ORDER BY length(window) DESC LIMIT 1
  RETURN length(longestChunkWindow)
  """

kg.query(cypher,
         params={'chunkIdParam': first_chunk_info['chunkId']})
````

### **`Neo4jVector.from_existing_graph()` **

`Neo4jVector.from_existing_graph()` 的輸出是一個 `Neo4jVector` 對象，該對象用於查詢和操作 Neo4j 圖數據庫中的向量數據，並通過檢索器 (`retriever`) 與 LLM 結合進行查詢增強。

---

### **輸出的數據內容**

#### **包含的數據**
1. **節點（Node）**
   - 從 Neo4j 資料庫中檢索到的節點（`Node`），包含與查詢匹配的節點屬性。
   - 屬性由初始化時指定的 `text_node_properties` 定義。

2. **分數（Score）**
   - 每個節點會帶有與查詢嵌入向量的相似性分數（通常基於餘弦相似度計算）。
   - 分數範圍：`[-1, 1]`，其中：
     - `1` 表示完全相似。
     - `-1` 表示完全不相似。

#### **返回的數量**
- 默認情況下，返回的節點數量依賴於檢索器的設置，通常為 **前 `k` 個**相似度最高的節點。
- 返回的數量可通過調整檢索器的參數進行控制。

---

### **調整參數**

可以通過以下方式調整返回的節點數量和數據內容：

#### **1. 調整檢索數量**
在 `neo4j_vector_store.as_retriever()` 中可以設置參數：
```python
windowless_retriever = neo4j_vector_store.as_retriever(search_k=5)
```
- **`search_k`**：
  - 返回與查詢最相似的前 `k` 個節點。
  - 默認值：`5`（可根據需求增加或減少）。

#### **2. 設置檢索屬性**
初始化 `Neo4jVector` 時設置以下參數：
- **`text_node_properties`**：
  - 定義要檢索的節點屬性，例如節點的 `text` 或其他描述性屬性。
  - 示例：
    ```python
    text_node_properties=["description", "content"]
    ```

- **`embedding_node_property`**：
  - 指定節點的嵌入向量屬性，用於計算相似度。

#### **3. 使用自定義查詢**
可以通過 `retrieval_query` 傳遞自定義 Cypher 查詢來控制返回數據：
```python
retrieval_query = """
MATCH (node:VectorNode)
WITH node, node.text AS text, node.score AS score
RETURN text, score, node {.source} AS metadata
"""
windowless_retriever = neo4j_vector_store.as_retriever(retrieval_query=retrieval_query)
```

- **`metadata`**：節點的其他屬性數據（由 `retrieval_query` 或初始化參數指定）。

### CONSTRAINT

---

### **程式片段 1: 使用 `MERGE`**

```cypher
MERGE (mgr:Manager {managerCik: $managerParam.managerCik})
  ON CREATE
      SET mgr.managerName = $managerParam.managerName,
          mgr.managerAddress = $managerParam.managerAddress
```

- **`MERGE` 的作用**：
  - `MERGE` 會檢查圖中是否已經存在符合條件的節點：
    - **條件**：節點的標籤是 `Manager`，且屬性 `managerCik` 等於 `$managerParam.managerCik`。
  - 如果存在，則返回該節點並跳過 `ON CREATE`。
  - 如果不存在，則創建新節點，並執行 `ON CREATE` 中定義的屬性設置。

- **潛在問題**：
  - 在沒有唯一性約束的情況下，**並發請求可能導致重複節點的創建**。
    - 例如，兩個 `MERGE` 查詢幾乎同時執行，而資料庫尚未完成第一次查詢的插入操作，導致第二個查詢認為該節點不存在，最終創建了重複的節點。

---

### **程式片段 2: 使用唯一性約束**

```cypher
CREATE CONSTRAINT unique_manager
  IF NOT EXISTS
  FOR (n:Manager)
  REQUIRE n.managerCik IS UNIQUE
```

- **唯一性約束的作用**：
  - 為 `Manager` 節點的屬性 `managerCik` 設置唯一性約束，保證資料庫層面上 `managerCik` 的值必須唯一。
  - 無論是手動執行的 `MERGE` 查詢還是其他查詢類型（如 `CREATE`），都不能插入具有重複 `managerCik` 的節點。
  - 如果嘗試插入重複的 `managerCik` 值，Neo4j 會直接拋出錯誤。

- **避免的問題**：
  - 無需依賴 `MERGE` 的邏輯處理，即使存在並發查詢，也能完全杜絕重複節點的生成。

---

### **兩者的不同之處**

| **特性**                        | **`MERGE` 的效果**                                              | **唯一性約束的效果**                                      |
|--------------------------------|----------------------------------------------------------------|---------------------------------------------------------|
| **重複節點**                    | 在並發情況下可能會創建重複節點                                | 保證資料庫層級不會有重複的節點，無需依賴程式邏輯         |
| **查詢依賴**                    | 依賴 `MERGE` 查詢來防止重複                                   | 對所有查詢生效，包括 `MERGE` 和 `CREATE`                |
| **性能影響**                    | 每次執行都需查詢是否存在匹配的節點                           | 初次創建約束可能需要時間，但後續操作性能更高            |
| **數據庫一致性**                | 無法完全保證一致性（在高併發情況下）                         | 強制性保證資料庫一致性，完全避免數據重複               |
| **錯誤處理**                    | 如果節點已存在，`MERGE` 不會報錯                              | 插入重複節點會拋出錯誤，需在應用邏輯中處理             |
| **適用範圍**                    | 只能在 `MERGE` 查詢中生效                                    | 對所有相關操作生效，包括外部直接操作資料庫的情況         |

---

### **推薦方式**

- 如果您需要在資料庫層級上完全避免重複節點，推薦使用 **唯一性約束**：
  ```cypher
  CREATE CONSTRAINT unique_manager
    IF NOT EXISTS
    FOR (n:Manager)
    REQUIRE n.managerCik IS UNIQUE
  ```
  這樣可以杜絕所有查詢引起的數據重複問題，並保證資料庫的一致性。

- **結合使用**：
  - 您可以繼續使用 `MERGE` 來處理插入或更新邏輯：
    ```cypher
    MERGE (mgr:Manager {managerCik: $managerParam.managerCik})
      ON CREATE
          SET mgr.managerName = $managerParam.managerName,
              mgr.managerAddress = $managerParam.managerAddress
    ```
  - 當配合唯一性約束使用時，`MERGE` 的查詢邏輯不會因為高併發而插入重複數據。

---

### **總結**

- **`MERGE`** 可以防止單次查詢中重複插入節點，但在高併發情況下，可能因競爭條件導致數據重複。
- **唯一性約束** 是最可靠的方法，在資料庫層級上保證屬性唯一，避免所有潛在重複問題。
- **最佳實踐** 是結合使用唯一性約束和 `MERGE`，以保證數據一致性和良好的查詢邏輯。

### VECTOR INDEX v.s. FULLTEXT INDEX v.s. INDEX

### **`CREATE VECTOR INDEX` 與 `CREATE FULLTEXT INDEX` 的區別**

#### **1. `CREATE VECTOR INDEX`**
- **用途**：
  - 用於支持向量搜索（Vector Search），即對高維嵌入數據進行相似性匹配。
  - 常用於語義搜索、圖像檢索、推薦系統等需要基於向量相似度進行匹配的場景。

- **工作方式**：
  - 針對節點的某個屬性（通常存儲為嵌入向量）創建向量索引。
  - 支持計算餘弦相似度、歐幾里得距離等，返回與查詢向量最相似的節點。

- **查詢方式**：
  - 使用 `CALL db.index.vector.queryNodes` 進行相似度查詢。

- **示例**：
  ```cypher
  CREATE VECTOR INDEX vectorIndexExample
    FOR (n:VectorNode)
    ON (n.embedding)
    OPTIONS {
      indexConfig: {
        `vector.dimensions`: 768,
        `vector.similarity_function`: 'cosine'
      }
    }
  ```

- **適用場景**：
  - 需要基於語義相似度（如文本、圖像、音頻等嵌入向量）進行檢索的應用。

---

#### **2. `CREATE FULLTEXT INDEX`**
- **用途**：
  - 用於全文檢索，支持對文本數據的模糊匹配和部分匹配。
  - 常用於檢索文檔、名稱、描述等結構化或非結構化的文本屬性。

- **工作方式**：
  - 為節點的文本屬性創建倒排索引，允許快速查找包含特定關鍵字的節點。

- **查詢方式**：
  - 使用 `CALL db.index.fulltext.queryNodes` 進行查詢。

- **示例**：
  ```cypher
  CREATE FULLTEXT INDEX fullTextIndexExample
    IF NOT EXISTS
    FOR (n:TextNode)
    ON EACH [n.text, n.description]
  ```

- **適用場景**：
  - 搜索包含特定單詞或短語的文本屬性，例如名稱查找、文檔檢索等。

---

### **主要區別**

| 特性                 | `CREATE VECTOR INDEX`                            | `CREATE FULLTEXT INDEX`                  |
|----------------------|------------------------------------------------|------------------------------------------|
| **目標數據**          | 高維嵌入向量                                    | 文本屬性                                 |
| **匹配方式**          | 基於數學相似度（如餘弦相似度、歐幾里得距離）     | 基於關鍵字模糊匹配或部分匹配             |
| **查詢場景**          | 語義搜索、推薦系統                              | 關鍵字檢索、文檔搜索                     |
| **索引結構**          | 基於向量的高效相似度索引（如 HNSW）             | 基於倒排索引                             |
| **查詢方式**          | `CALL db.index.vector.queryNodes`               | `CALL db.index.fulltext.queryNodes`      |
| **適用場景**          | 需要語義檢索或基於向量匹配的場景                | 需要基於關鍵詞或文本內容進行檢索的場景   |

---

### **其他索引類型**

#### **1. 屬性索引**
- **描述**：
  - 針對節點或關係的屬性進行索引，用於加速精確匹配查詢（如 `=`、`IN`）。
- **用途**：
  - 適合結構化數據的檢索，如查找特定屬性值的節點。
- **示例**：
  ```cypher
  CREATE INDEX propertyIndexExample
    IF NOT EXISTS
    FOR (n:Node)
    ON (n.property)
  ```
- **查詢方式**：
  - 常用於 `MATCH` 查詢中，例如：
    ```cypher
    MATCH (n:Node {property: "value"})
    RETURN n
    ```

---

#### **2. 唯一性約束**
- **描述**：
  - 為節點的某個屬性設置唯一性約束，確保該屬性的值在圖中是唯一的。
- **用途**：
  - 適合用於確保節點的唯一標識符（如 ID、CUID）不重複。
- **示例**：
  ```cypher
  CREATE CONSTRAINT uniqueConstraintExample
    IF NOT EXISTS
    FOR (n:Node)
    REQUIRE n.uniqueProperty IS UNIQUE
  ```

---

#### **3. 範圍索引**
- **描述**：
  - 用於加速範圍查詢（如數字或日期）。
- **用途**：
  - 適合數值型數據的篩選，例如查找特定日期範圍內的節點。
- **示例**：
  ```cypher
  CREATE INDEX rangeIndexExample
    IF NOT EXISTS
    FOR (n:Node)
    ON (n.dateProperty)
  ```
- **查詢方式**：
  - 用於範圍查詢，例如：
    ```cypher
    MATCH (n:Node)
    WHERE n.dateProperty > date("2023-01-01")
    RETURN n
    ```

---

### **索引類型總覽與選擇**

| **索引類型**         | **用途**                                 | **查詢場景**                               |
|----------------------|------------------------------------------|--------------------------------------------|
| `VECTOR INDEX`       | 向量檢索                                 | 語義搜索、推薦系統                         |
| `FULLTEXT INDEX`     | 文本檢索                                 | 文檔搜索、名稱匹配                         |
| `PROPERTY INDEX`     | 精確匹配                                 | 節點屬性查詢（`=`、`IN`）                  |
| 唯一性約束            | 保證屬性唯一性                           | 節點或關係的唯一標識符檢查                 |
| 範圍索引             | 加速範圍查詢                             | 日期範圍、數字範圍查詢                     |

---

### **總結**
1. **`VECTOR INDEX`**：處理嵌入向量的高維相似性檢索，適合語義搜索。
2. **`FULLTEXT INDEX`**：用於關鍵字匹配的全文檢索，適合文本查詢。
3. **其他索引**：如屬性索引、唯一性約束和範圍索引，分別適合結構化數據和特殊查詢需求。
4. **選擇索引**：根據查詢場景選擇合適的索引類型，提升查詢效率和結果準確性。

在 Neo4j 中，範圍索引和精確匹配索引雖然在功能上有不同的用途（範圍查詢 vs 精確匹配），但在 **命令語法層面上**，它們的創建方式是一致的，都是通過 `CREATE INDEX` 指令來實現。

---

### **索引方式的差異**

#### **1. 精確匹配索引**
- **用途**：
  - 用於加速精確匹配查詢，例如 `=` 或 `IN`。
  - 適合用於處理具有明確標識符或離散值的數據（如 ID 或枚舉類型屬性）。

- **索引實現**：
  - 在底層，Neo4j 通常使用 **哈希索引** 來支持精確匹配。
  - 哈希索引在處理範圍查詢時效率較低，因此不適合用於處理連續數據（如數字或日期）。

#### **2. 範圍索引**
- **用途**：
  - 用於加速範圍查詢，例如 `<`、`>`、`BETWEEN`。
  - 適合處理連續型數據（如數字、日期、浮點數等）。

- **索引實現**：
  - 在底層，Neo4j 使用 **B+樹索引** 來支持範圍查詢。
  - B+樹索引能高效地處理範圍操作，支持按序查詢和排序。

---

### **在命令行創建方式相同**

無論是精確匹配索引還是範圍索引，**創建索引的命令格式相同**，均為：
```cypher
CREATE INDEX index_name
IF NOT EXISTS
FOR (node_label)
ON (node_label.property)
```

#### **範例 1：精確匹配索引**
適合對離散值的精確查詢，例如 ID 或固定類型的標籤：
```cypher
CREATE INDEX exactMatchIndex
IF NOT EXISTS
FOR (n:User)
ON (n.userId)
```
查詢：
```cypher
MATCH (n:User {userId: "12345"})
RETURN n
```

#### **範例 2：範圍索引**
適合對連續值進行範圍查詢，例如日期或數值：
```cypher
CREATE INDEX rangeIndex
IF NOT EXISTS
FOR (n:Event)
ON (n.eventDate)
```
查詢：
```cypher
MATCH (n:Event)
WHERE n.eventDate > date("2023-01-01")
RETURN n
```

---

### **如何區分索引類型**

1. **自動區分**：
   - Neo4j 會根據索引屬性數據的類型和查詢模式，選擇適合的索引結構。
   - 對於數字或日期屬性，Neo4j 預設使用範圍索引（B+樹），以便支持範圍查詢。
   - 對於其他離散類型數據（如字符串或標識符），使用哈希索引以支持精確匹配。

### CALL db.index

在 Neo4j 中，`CALL db.index` 提供了多種與索引操作相關的查詢和存儲過程，用於管理和查詢索引。以下是一些常用的 `CALL db.index` 方法及其用途：

---

### **1. 索引查詢相關方法**

#### **`db.index.fulltext.queryNodes`**
- **用途**：基於全文索引（`FULLTEXT INDEX`）進行節點的全文檢索。
- **語法**：
  ```cypher
  CALL db.index.fulltext.queryNodes("indexName", "searchString")
  YIELD node, score
  RETURN node, score
  ```
- **參數**：
  - `indexName`：全文索引的名稱。
  - `searchString`：檢索字符串（支持關鍵字查詢）。
- **返回值**：
  - `node`：匹配的節點。
  - `score`：相似度分數。
- **示例**：
  ```cypher
  CALL db.index.fulltext.queryNodes("fullTextManagerNames", "royal bank")
  YIELD node, score
  RETURN node.managerName, score
  ```

#### **`db.index.fulltext.queryRelationships`**
- **用途**：基於全文索引進行關係的全文檢索。
- **語法**：
  ```cypher
  CALL db.index.fulltext.queryRelationships("indexName", "searchString")
  YIELD relationship, score
  RETURN relationship, score
  ```
- **示例**：
  ```cypher
  CALL db.index.fulltext.queryRelationships("fullTextRelationshipIndex", "owns")
  YIELD relationship, score
  RETURN relationship, score
  ```

#### **`db.index.vector.queryNodes`**
- **用途**：基於向量索引（`VECTOR INDEX`）進行節點的相似度檢索。
- **語法**：
  ```cypher
  CALL db.index.vector.queryNodes("indexName", topK, queryVector)
  YIELD node, score
  RETURN node, score
  ```
- **參數**：
  - `indexName`：向量索引名稱。
  - `topK`：返回與查詢向量最相似的節點數量。
  - `queryVector`：查詢向量（通常是嵌入向量）。
- **示例**：
  ```cypher
  CALL db.index.vector.queryNodes("vectorIndexExample", 5, [0.1, 0.2, 0.3])
  YIELD node, score
  RETURN node, score
  ```

---

### **2. 索引管理相關方法**

#### **`SHOW INDEXES`**
- **用途**：列出資料庫中所有的索引及其狀態。
- **語法**：
  ```cypher
  SHOW INDEXES
  ```
- **返回值**：
  - 索引的描述、類型、應用的屬性等。


#### **`db.index.fulltext.createNodeIndex`**
- **用途**：通過 Cypher 創建節點的全文索引。
- **語法**：
  ```cypher
  CALL db.index.fulltext.createNodeIndex("indexName", ["NodeLabel"], ["property1", "property2"])
  ```
- **參數**：
  - `indexName`：索引名稱。
  - `NodeLabel`：要應用的節點標籤。
  - `property1`, `property2`：要索引的屬性。
- **示例**：
  ```cypher
  CALL db.index.fulltext.createNodeIndex("fullTextManagerNames", ["Manager"], ["managerName"])
  ```

#### **`db.index.fulltext.createRelationshipIndex`**
- **用途**：通過 Cypher 創建關係的全文索引。
- **語法**：
  ```cypher
  CALL db.index.fulltext.createRelationshipIndex("indexName", ["RelationshipType"], ["property1", "property2"])
  ```

在這段 Cypher 查詢中，`RETURN owns { .shares, .value }` 是使用了一種稱為 **屬性投影（property projection）** 的語法，用於從關係 `owns` 中選擇特定的屬性進行返回。

---

1. **選擇性返回屬性**：
   - `{ .shares, .value }` 表示從關係 `owns` 中提取 `shares` 和 `value` 這兩個屬性。
   - 這樣可以只返回感興趣的屬性，而不是返回整個關係的所有屬性。

2. **返回結構化數據**：
   - 投影的屬性會被封裝為一個字典格式的結構，讓返回的結果更加清晰且易於使用。
   - 例如，返回的結果可能是：
     ```json
     {
       "shares": 1000,
       "value": 50000
     }
     ```

3. **提高性能**：
   - 如果關係中有很多屬性，但您只關心其中的幾個，使用屬性投影可以減少數據傳輸量和處理成本。

````python
investment_retrieval_query = """
MATCH (node)-[:PART_OF]->(f:Form),
    (f)<-[:FILED]-(com:Company),
    (com)<-[owns:OWNS_STOCK_IN]-(mgr:Manager)
WITH node, score, mgr, owns, com
    ORDER BY owns.shares DESC LIMIT 10
WITH collect (
    mgr.managerName +
    " owns " + owns.shares +
    " shares in " + com.companyName +
    " at a value of $" +
    apoc.number.format(toInteger(owns.value)) + "."
) AS investment_statements, node, score
RETURN apoc.text.join(investment_statements, "\n") +
    "\n" + node.text AS text,
    score,
    {
      source: node.source
    } as metadata
"""
````

在 Cypher 中，WITH 子句用於將查詢的中間結果傳遞給後續的查詢部分，並可以對這些中間結果進行操作，如排序、分組或限制返回的數據量。

### integrate_info_into_kg

````python
def integrate_info_into_kg(info_cap, kg):
    """
    将 info_cap 的信息整合进 Neo4j 图 kg。
    """
    for record in info_cap:
        # 获取 Company 和 Address 信息
        company = record['co']
        address = record['address']

        # 将信息整合到图中
        kg.query("""
        MERGE (co:Company {cusip: $company.cusip})
          ON CREATE SET co.companyName = $company.companyName,
                        co.cusip6 = $company.cusip6,
                        co.companyAddress = $company.companyAddress,
                        co.location = point($company.location),
                        co.names = $company.names
          ON MATCH SET co.companyName = coalesce(co.companyName, $company.companyName),
                       co.companyAddress = coalesce(co.companyAddress, $company.companyAddress),
                       co.location = coalesce(co.location, point($company.location)),
                       co.names = coalesce(co.names, $company.names)

        MERGE (addr:Address {location: point($address.location)})
          ON CREATE SET addr.country = $address.country,
                        addr.city = $address.city,
                        addr.state = $address.state
          ON MATCH SET addr.country = coalesce(addr.country, $address.country),
                       addr.city = coalesce(addr.city, $address.city),
                       addr.state = coalesce(addr.state, $address.state)

        MERGE (co)-[:LOCATED_AT]->(addr)
        """, {
            "company": {
                "cusip": company['cusip'],
                "cusip6": company['cusip6'],
                "companyName": company['companyName'],
                "companyAddress": company['companyAddress'],
                "location": {
                    "longitude": company['location'][0],  # 使用索引访问经度
                    "latitude": company['location'][1]   # 使用索引访问纬度
                },
                "names": company['names']
            },
            "address": {
                "country": address['country'],
                "city": address['city'],
                "state": address['state'],
                "location": {
                    "longitude": address['location'][0],  # 使用索引访问经度
                    "latitude": address['location'][1]   # 使用索引访问纬度
                }
            }
        })
````

## Node & Relationship

在 **Neo4j** 中，數據的基礎單位是 **節點 (Node)** 和 **關係 (Relationship)**，這兩者共同構成 **圖數據庫** 的結構。以下是節點和關係所能攜帶的所有資訊，以及常見名詞對照和必要性分類。

---

## **1. 節點 (Node)**

### **節點能攜帶的資訊：**
| 資訊項        | 英文名稱            | 說明                           | 必要性   | 備註                      |
|---------------|--------------------|--------------------------------|---------|---------------------------|
| **標籤**      | **Label**          | 節點的分類標識。標籤可以有多個。   | **必要** | 節點至少需要一個標籤，便於分類。|
| **屬性**      | **Properties**     | 鍵值對 (key-value pair)，附加的資訊。 | **可選** | 包含節點的詳細數據，例如名稱、ID、年齡等。|
| **唯一標識符**| **Internal ID**    | 系統生成的唯一識別碼 (自動生成)。 | **必有唯一** | 不需要手動設置，自動生成。|

---

### **常見名詞對照：**
- **節點 (Node)**：也常被稱為 **點** 或 **Vertex**。
- **標籤 (Label)**：表示節點的類型或分類，例如 `Person`、`Movie`。
- **屬性 (Properties)**：附加於節點的詳細數據。
- **唯一標識符 (Internal ID)**：每個節點在 Neo4j 中的唯一識別碼。

---

### **節點例子**
```cypher
CREATE (:Person {name: "Keanu Reeves", age: 58})
CREATE (:Movie {title: "The Matrix", released: 1999})
```
- **`Person` 和 `Movie`** 是標籤 (Labels)。
- **`name`、`age` 和 `title`、`released`** 是屬性 (Properties)。


在 Neo4j 中，節點可以同時擁有多個標籤，用冒號 ： 分隔標籤即可。

```cypher
CREATE (:Person:Actor {name: "Keanu Reeves", age: 58})
```


---

## **2. 關係 (Relationship)**

### **關係能攜帶的資訊：**
| 資訊項          | 英文名稱            | 說明                           | 必要性   | 備註                      |
|-----------------|--------------------|--------------------------------|---------|---------------------------|
| **類型**        | **Type**           | 關係的標識，例如 `ACTED_IN`。     | **必有唯一** | 每個關係只有一個類型而且必須有一個類型。    |
| **方向**        | **Direction**      | 關係的方向性：從節點 A 指向節點 B。 | **必要** | 必須有方向性，但可以設置為雙向。 |
| **屬性**        | **Properties**     | 鍵值對 (key-value pair)，附加的資訊。 | **可選** | 包含關係的詳細數據，例如角色名稱、關係權重等。 |
| **唯一標識符**  | **Internal ID**    | 系統生成的唯一識別碼 (自動生成)。 | **必有唯一** | 不需要手動設置，自動生成。|

---

### **常見名詞對照：**
- **關係 (Relationship)**：也常被稱為 **邊 (Edge)**。
- **類型 (Type)**：關係的標識，例如 `ACTED_IN`、`DIRECTED`。
- **方向 (Direction)**：關係有方向性，如從節點 A 到節點 B。
- **屬性 (Properties)**：附加於關係的詳細數據。
- **唯一標識符 (Internal ID)**：每個關係的唯一識別碼。

---

### **關係例子**
```cypher
CREATE (:Person {name: "Keanu Reeves"})-[:ACTED_IN {role: "Neo"}]->(:Movie {title: "The Matrix"})
```
- **`:ACTED_IN`** 是關係的**類型 (Type)**。
- **`role: "Neo"`** 是關係的**屬性 (Properties)**。
- 關係的方向是從 **`Person`** 指向 **`Movie`**。

---

## **3. 節點與關係資訊的必要性**

| 資訊類型        | 必要性         | 說明                                     |
|-----------------|---------------|------------------------------------------|
| **節點標籤 (Label)**       | 必要          | 節點至少有一個標籤，標識節點的類別。           |
| **節點屬性 (Properties)**  | 可選          | 可以不設置，但通常會用來存儲節點的詳細數據。    |
| **關係類型 (Type)**        | 必要          | 關係必須有類型，標識關係的含義。              |
| **關係屬性 (Properties)**  | 可選          | 可以不設置，但通常用來存儲關係的詳細數據。      |
| **方向 (Direction)**       | 必要          | 關係的方向必須被明確設置，從一個節點指向另一個。|

---

## **4. 總結與圖示**

### **節點：**
- **必須**：標籤 (Label)、唯一標識符 (Internal ID)。
- **可選**：屬性 (Properties)。

### **關係：**
- **必須**：類型 (Type)、方向 (Direction)、唯一標識符 (Internal ID)。
- **可選**：屬性 (Properties)。

---

### **圖形結構的視覺化：**
一個完整的例子：

```plaintext
(:Person {name: "Keanu Reeves"}) -[:ACTED_IN {role: "Neo"}]-> (:Movie {title: "The Matrix"})
```

- **節點**：
  - `:Person` 和 `:Movie` 是節點標籤。
  - `name` 和 `title` 是節點的屬性。
- **關係**：
  - `:ACTED_IN` 是關係的類型。
  - `{role: "Neo"}` 是關係的屬性。
  - 關係方向：從 `Person` 節點指向 `Movie` 節點。

---

### **附加常用名詞對照表**
| 英文名詞      | 中文翻譯         | 常用名詞        |
|---------------|-----------------|-----------------|
| Node          | 節點            | 點 (Vertex)     |
| Relationship  | 關係            | 邊 (Edge)       |
| Label         | 標籤            | 類別            |
| Type          | 類型 (關係)     | 類型            |
| Property      | 屬性            | 屬性資料        |
| Direction     | 方向            | 邊的方向        |
| Internal ID   | 唯一標識符      | 自動生成的 ID   |

````python
import re
from neo4j.spatial import Point

def convert_text_to_dict_safe(text):
    """
    将文本字符串转换为字典，并将 POINT() 转换为 Point()。
    """
    # 替换 POINT(-x y) 为 Point((-x, y)) 的正则表达式
    point_pattern = re.compile(r"POINT\((-?\d+\.?\d*)\s+(-?\d+\.?\d*)\)")

    # 替换文本中的 POINT 为 Point 格式
    converted_text = point_pattern.sub(r"Point((\1, \2))", text)

    # 使用 eval 将文本解析为 Python 对象
    try:
        parsed_dict = eval(converted_text, {"Point": Point})
    except Exception as e:
        raise ValueError(f"无法解析文本为字典: {e}")

    return parsed_dict

# 原始文本字符串
text = """
[
    {
        'co': {
            'cusip': '697435105',
            'names': ['Palo Alto Networks Inc.', 'PALO ALTO NETWORKS INC', 'PALO ALTO NETWORKS INC PUT', 'None'],
            'companyAddress': '3000 Tannery Way, Santa Clara, CA 95054, USA',
            'companyName': 'PALO ALTO NETWORKS INC',
            'cusip6': '697435',
            'location': POINT(-121.9828954 37.383283)
        },
        'address': {
            'country': 'United States',
            'city': 'Santa Clara',
            'state': 'California',
            'location': Point((-121.9323029, 37.3473759))  # 使用元组
        }
    },
    {
        'co': {
            'cusip': '31428X106',
            'names': ['FedEx Corp', 'FEDEX CORP'],
            'companyAddress': 'Delaware, USA',
            'companyName': 'FEDEX CORP',
            'cusip6': '31428X',
            'location': Point((-75.52766989999999, 38.9108325))  # 使用元组
        },
        'address': {
            'country': 'United States',
            'city': 'Delaware',
            'state': 'Delaware',
            'location': POINT(-75.52766989999999 38.9108325)
        }
    }
]
"""

# 转换文本为字典
info_cap = convert_text_to_dict_safe(text)

# 输出结果
print("转换后的字典:")
print(info_cap)
````

### integrate_mgrinfo_into_kg

````python
def integrate_mgrinfo_into_kg(info_cap, kg):
    """
    将 info_cap 的 Manager 信息整合进 Neo4j 图 kg。
    """
    for record in info_cap:
        # 获取 Manager 和 Address 信息
        manager = record['mgr']
        address = record['address']

        # 将信息整合到图中
        kg.query("""
        MERGE (mgr:Manager {managerCik: $manager.managerCik})
          ON CREATE SET mgr.managerName = $manager.managerName,
                        mgr.managerAddress = $manager.managerAddress,
                        mgr.location = point($manager.location)
          ON MATCH SET mgr.managerName = coalesce(mgr.managerName, $manager.managerName),
                       mgr.managerAddress = coalesce(mgr.managerAddress, $manager.managerAddress),
                       mgr.location = coalesce(mgr.location, point($manager.location))

        MERGE (addr:Address {location: point($address.location)})
          ON CREATE SET addr.country = $address.country,
                        addr.city = $address.city,
                        addr.state = $address.state
          ON MATCH SET addr.country = coalesce(addr.country, $address.country),
                       addr.city = coalesce(addr.city, $address.city),
                       addr.state = coalesce(addr.state, $address.state)

        MERGE (mgr)-[:LOCATED_AT]->(addr)
        """, {
            "manager": {
                "managerCik": manager['managerCik'],
                "managerName": manager['managerName'],
                "managerAddress": manager['managerAddress'],
                "location": {
                    "longitude": manager['location'][0],  # 使用索引访问经度
                    "latitude": manager['location'][1]   # 使用索引访问纬度
                }
            },
            "address": {
                "country": address['country'],
                "city": address['city'],
                "state": address['state'],
                "location": {
                    "longitude": address['location'][0],  # 使用索引访问经度
                    "latitude": address['location'][1]   # 使用索引访问纬度
                }
            }
        })


# 将数据整合到图中
integrate_mgrinfo_into_kg(info_cap, kg)
````

````python
import ast
import re
from neo4j.spatial import Point

def convert_text_to_dict_safe2(text):
    """
    Converts a text string containing POINT(x y) to a Python dictionary,
    replacing POINT(x y) with Point((x, y)).
    """

    def replace_point(match):
        # Extract the numbers inside POINT(x y)
        coordinates = match.group(1).split()
        return f"Point(({coordinates[0]}, {coordinates[1]}))"

    # Debug: Log the original text
    print("Original text:", text[:500], "...")  # Only show the first 500 characters for clarity

    # Replace POINT(x y) with Point((x, y))
    adjusted_text = re.sub(r"POINT\((-?\d+\.\d+ -?\d+\.\d+)\)", replace_point, text)

    # Debug: Log the adjusted text
    print("Adjusted text:", adjusted_text[:500], "...")

    # Ensure the adjusted text has valid syntax
    try:
        parsed_dict = ast.literal_eval(adjusted_text)
    except SyntaxError as e:
        print("Syntax error in adjusted text:", adjusted_text[:500], "...")
        raise ValueError(f"Failed to parse text into dictionary: {e}")
    except Exception as e:
        print("Error evaluating adjusted text:", adjusted_text[:500], "...")
        raise ValueError(f"Failed to parse text into dictionary: {e}")

    return parsed_dict


# Convert text to dictionary
try:
    info_cap = convert_text_to_dict_safe2(text)
    print("Converted dictionary:", info_cap)
except ValueError as e:
    print("Conversion failed:", e)



````
